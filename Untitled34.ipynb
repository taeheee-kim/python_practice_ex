```python
file = open("C:/Users/kth25/Desktop/file2read.txt", 'r')
line = file.read()
```

```python
print(line)
```

```python
word = line.replace("\n", ' ')
word = word.replace('.', ' ')
word = word.split()
```

```python
word
```

```python
wordcount = {}
```

```python
for i in word:
    print('단어 : '+i)
    print('이전까지 해당 단어 사용 횟수 : 'wordcount.get(i,0))
    wordcount[i] = wordcount.get(i, 0)+1
    print('1번 더 count 반영한 단어 사용 횟수 : 'wordcount[i])
    print('지금까지 등장한 단어 : 'wordcount.keys())
    print(sorted(wordcount.keys()))
    keys = sorted(wordcount.keys())
```
